# Apiriori recommendation


```python
basket = dataset.groupby(["transaction","product"])["product"].count().reset_index(name="count")
basket.head()
```

```python
basket = basket.pivot_table(index="transaction",columns="product",values="count")
basket.head()
```

```python
basket = basket.fillna(0)
basket.head()
```

```python
def decode(x):
    if x>0:
        return 1
    else:
        return 0
```

```python
basket = basket.applymap(decode)
basket.head()
```

```python
from mlxtend.frequent_patterns import association_rules, apriori 
```

```python
friquent_item = apriori(basket,min_support=0.05,use_colnames=True)
friquent_item
```

```python
rules = association_rules(friquent_item, metric="lift",min_threshold=1, num_itemsets=1)
rules
```

```python
rules[rules["antecedents"]=={"Latte"}].sort_values("confidence",ascending=False)
```

```python
catagories = dataset[["product","product_category"]].drop_duplicates().set_index("product").to_dict()['product_category']
catagories
```

```python
catagories.get('Cappuccino')
```

```python
recommendation_json = {}

antecedents = rules["antecedents"].unique()

for antecedent in antecedents:
    dff = rules[(rules["antecedents"] == antecedent)]
    dff = dff.sort_values("confidence",ascending=False)

    key = "_".join(antecedent)
    recommendation_json[key] = []

    for _,row in dff.iterrows():
        objs = rules["consequents"]
        for obj in objs:
            ob = next(iter(obj))
            rec = {
                "product": ob,
                "product_category": catagories.get(ob),
                "confidence": row["confidence"]
            }

            recommendation_json[key].append(rec)

        

        
```
